```python
import os
import sys
import glob
from pathlib import Path
sys.path.append("src")

from qscore import score
```

```python
#### Important
###### For this step, please:
###### - Install ChimeraX and open it
###### - On menu, press "Tools" -> "More Tools..."
###### - Search for "score" and install "QScore"
###### - Find the binary path for ChimeraX
```

```python
# Path settings
root_dir     = Path('./data')
chimera_path = '/Applications/ChimeraX_Daily.app/Contents/MacOS/ChimeraX'
pdb_ext      = '.cif'
volume_exts  = ['.mrc']
```

```python
qpaths = score(
    root_dir=root_dir,
    chimera_path=chimera_path,
    pdb_ext=pdb_ext,
    volume_exts=volume_exts
)
```

```python
# Now, copy QScores to the prealigned folder
target = "./data/aligned" # Output path for aligned PDBs
for q in qpaths: os.system(f"cp {q} {target}")
```
